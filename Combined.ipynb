```python
# Import dependencies
import pandas as pd
import requests
import json
import time

from config import client_id_spot, client_secret_spot, api_key_last, secret_last

import spotipy
import spotipy.util as util
from spotipy import Spotify
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
# Read HookTheory data and store in dataframe
hook_data = pd.read_csv('chords_all_songs.csv')

song_data = hook_data[['song','artist','cp','section']]
song_data.head()
```

```python
url_last = 'http://ws.audioscrobbler.com/2.0/?'

endpoint_last = 'method=track.getInfo'

#song_data['Track (last.fm)'] = ''
#song_data['Track MBID (last.fm)'] = ''
#song_data['Artist (last.fm)'] = ''
#song_data['Artist MBID (last.fm)'] = ''
#song_data['Listener Count (last.fm)'] = ''
#song_data['Play Count (last.fm)'] = ''
#song_data['Genre (last.fm)'] = ''

#for i in range (0,1):
for i in range(len(song_data)):
    
    if (i>0) & (i % 100 == 0):
        print('Waiting 60 seconds before resuming request...')
        time.sleep(20)
    
    else:
        artist = song_data.loc[i,'artist']
        track = song_data.loc[i,'song']
    
        param_last = {'api_key':api_key_last,
                      'artist':artist,
                      'track':track,
                      'format':'json'}
    
        response_last = requests.get(url_last+endpoint_last, params=param_last).json()
    
        print(f'Retrieving information for "{track}" by {artist}...')
    
        try:
            song_data.loc[i,'Track (last.fm)'] = response_last['track']['name']
            #song_data.loc[i,'MBID (last.fm)'] = response_last['track']['mbid']
            song_data.loc[i,'Artist (last.fm)'] = response_last['track']['artist']['name']
            #song_data.loc[i,'Artist MBID (last.fm)'] = response_last['track']['artist']['mbid']
            song_data.loc[i,'Listener Count (last.fm)'] = response_last['track']['listeners']
            song_data.loc[i,'Play Count (last.fm)'] = response_last['track']['playcount']
        
            tags = ''
        
            for j in range(len(response_last['track']['toptags']['tag'])):
                if j == 0:
                    tags = tags + response_last['track']['toptags']['tag'][j]['name']
                else:
                    tags = f"{tags}, {response_last['track']['toptags']['tag'][j]['name']}"
        
            song_data.loc[i,'Genre (last.fm)'] = tags
    
        except:
            print("\tTrack not found...")
```

```python
# Request access token from Spotify API
token = SpotifyClientCredentials(client_id=client_id_spot,
                                 client_secret=client_secret_spot)
cache_token = token.get_access_token()
```

```python
url_spot = 'https://api.spotify.com/v1'
endpoint_spot = '/search'

# Store access token as a header variable
header = {'Authorization':f'Bearer {cache_token}'}

#song_data['Track (Spotify)'] = ''
#song_data['Artist (Spotify)'] = ''
#song_data['Popularity (Spotify)'] = ''

#for i in range (0,1):
for i in range(len(song_data)):
    if (i>0) & (i % 100 == 0):
        print('Waiting 60 seconds before resuming request...')
        time.sleep(20)
    else:
        track =  song_data.loc[i,'Track (last.fm)']
        artist = song_data.loc[i,'Artist (last.fm)']
    
        param_spot = {'q':track,
                      'type':'track',
                      'artist':artist,
                      'limit':50}

        print(f'Retrieving information for "{track}" by {artist}...')
    
        response_spot = requests.get(url_spot+endpoint_spot, headers=header, params=param_spot).json()
    
        try:
            for j in range(len(response_spot['tracks']['items'])):
                track_spot = response_spot['tracks']['items'][j]['name']
                artist_spot = response_spot['tracks']['items'][j]['artists'][0]['name']
            
                if (track == track_spot) & (artist == artist_spot):
                    song_data.loc[i,'Track (Spotify)'] = track_spot
                    song_data.loc[i,'Artist (Spotify)'] = artist_spot
                    song_data.loc[i,'Popularity (Spotify)'] = response_spot['tracks']['items'][j]['popularity']
                    song_data.loc[i,'Release Date (Spotify)'] = response_spot['tracks']['items'][j]['album']['release_date']
                    break
        except:
            print('\tTrack not found...')
```
