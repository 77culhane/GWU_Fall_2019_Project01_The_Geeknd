```python
# Import dependencies
import pandas as pd
```

```python
# Import API data
hook_data = pd.read_csv('chords_all_songs.csv')
hook_data = hook_data.drop(columns='Unnamed: 0')
hook_data.head()
```

```python
hook_data = hook_data.sort_values(by=['artist','song'])
hook_data.head(50)
```

```python
hook_data = hook_data.drop_duplicates(subset=['artist','song'])
hook_data.head(50)
```

```python
hook_data.to_csv('unique_songs.csv',index=False)
```
