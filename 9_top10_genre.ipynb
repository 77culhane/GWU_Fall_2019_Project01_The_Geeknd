```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import time
import numpy
```

```python
# Save resulting dataframe as a csv file
combined_all = pd.read_csv('combined_all.csv')
combined_all.head(1)
```

```python
list(combined_all['genre (last.fm)'].unique())
```

```python
# Create dataframe for target genre
pop_songs = combined_all.loc[combined_all['genre (last.fm)'] == 'pop']

# Sort resulting dataframe based on artist, song and chord type in descending order
pop_songs = pop_songs.sort_values(['artist','song','type'], ascending=False)
pop_songs
```

```python
# Drop duplicate entries using artist and song values as criteria
pop_songs_unique = pop_songs.drop_duplicates(subset=['artist','song'])

# Sort resulting dataframe based on chord progression and popularity on Spotify
pop_songs_unique = pop_songs_unique.sort_values(['cp','popularity (Spotify)'],ascending=False)
pop_songs_unique
```

```python
# Group data by chord progression and aggregate using count of chord progression and maximum popularity value
pop = pop_songs_unique.groupby(['cp']).agg({'cp':'count','popularity (Spotify)':'max'})
pop.columns = ['count','max popularity']
pop = pop.reset_index()
pop = pop.sort_values('count',ascending=False)
pop = pop.reset_index(drop=True)
pop
```

```python
# Display number of unique chord progressions
len(list(pop_songs_unique['cp'].unique()))
```

```python
# Plot values for top 10 chord progressions
pop_top10 = [pop.loc[_,'cp'] for _ in range(10)]
pop_top10

pop_top10_count = [pop.loc[_,'count'] for _ in range(10)]
pop_top10_count

plt.bar(pop_top10, pop_top10_count)
plt.title('Most Popular Chord Progressions for Pop Songs')
plt.xlabel('Chord Progression')
plt.ylabel('Song Count')
plt.xticks(rotation='45')

plt.show()
```
